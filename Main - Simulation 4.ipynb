* Problem Statement:A couple plans to have children until they have a girl or until they have four children, whichever comes first.Estimate the likelihood that they will have a girl. Assume that each child has probability 0.5 of being a girl and 0.5 of being a boy, and the sexes of successive children are independent.

### Single Trial Setup

* Generate a random integer between 1 and 100
 * #'s 1-50 represent couple having a girl
      * 50% chance
 * #'s 51-100 represent couple having a boy
      * 50% chance

```python
# Write code to run a single trial
# Refer to your World Series Simulation as a resource!

# import the numpy package
import numpy as np

# Generate random number between 1-100
single_child = np.random.randint(1,101)

# Test the percentage ranges
if single_child <= 50:
    print("girl")
else:
    print("boy")
```

### Single Trial Function
* Return a String of "girl" or "boy"
* Abstraction

```python
# Define a function that runs a single trial and returns its result appropriately
def single_child_born():
    # Generate random number between 1-100
    single_child = np.random.randint(1,101)

    # Test the percentage ranges
    # Returning a string of a girl.boy child born
    if single_child <= 50:
        return("girl")
    else:
        return("boy")   

print(single_child_born())
```

### Simulate a single trial until the couple has a girl or has four children total
* Create a variable that tracks the total number of children, the number of boys, and the number of girls the couple has.
    * Start at 0!
* Utilize a "while" loop
     * While your single day trial returns that number of girls couple has is zero AND total number of children < 4,add one to the total number of children the couple has.
* Use if & else statement
    * If the single child born is a girl, add one to number of girls couple has and single child born is a boy, add one to number of boys.
* Print the variable tracking the total number of children the couple has.

```python
# import the numpy package
import numpy as np

# Create variables
total_num_of_child = 0
num_boys = 0
num_girls = 0

# While Loop
while(num_girls == 0 and total_num_of_child < 4):
    total_num_of_child += 1
    single_child = single_child_born()
    if (single_child == "girl"):
            num_girls += 1
    else:
            num_boys += 1
    
# Print results
print(total_num_of_child)
```

#### Copy and paste code into a function
* Return the total number of children the couple has
* Abstraction!

```python
# Create a function that simulates a complete single trial
def total_num_of_child_trial():
    total_num_of_child = 0
    num_boys = 0
    num_girls = 0
    
    while (num_girls == 0 and total_num_of_child < 4):
        total_num_of_child = total_num_of_child + 1
        single_child = single_child_born()
        if (single_child == "girl"):
            num_girls += 1
        else:
            num_boys += 1
    return total_num_of_child
print(total_num_of_child_trial())
```

### Full Simulation: finding the likelihood of the couple having a girl
* Create an array to hold your results
* Create a loop to run 10000 times
    * Add result of each total number of children trial to the array

```python
# Run your single trial a large number of times
# Record your data in an array

# Create variables
total_num_of_child_results = []

# Create a loop that represents a large number of trials(10000 trials in this case)
for i in range(10000):
    result = total_num_of_child_trial()
    total_num_of_child_results.append(result)

# Print the results
print(total_num_of_child_results)
```

### Calculate the Results
* Finding the total number of times the couple had a girl in the 10000 trials before the couple had four children.
* Calculating the probability of the couple having a girl before having four children

```python
# Using the results array, grab appropriate data and perform the necessary calculations
total_trials = 10000
girl = 0
boy = 0

for trial in total_num_of_child_results:
    if trial < 4:
        girl += 1
    else:
        boy += 1
print("Girls: " + str(girl))
print("Four children: " + str(boy))
```

```python
# Calculate probability
probability = girl / total_trials

# Print results
print(probability)
```

### Display the Results
* Pie chart

```python
# Graphically display results related directly to your question
# Import the library
import matplotlib.pyplot as plot
%matplotlib inline
# Magic to allow the graph to display directly in this notebook

# Create an array of labels
labels = ["Couple has Girl first", "Couple Has Four Children first"]

# Create an array of your results
result_bin = [girl, boy]

# Explode option
    # 'Slices' appear distanced from the center
        # Larger numbers = further explosion
    # Explode array should be same size as labels
explode = (0, 0.2)

# Use matplotlib module subplots() to get data for various charts
    # Returns a tuple in the form (figure,axes)
fig1,ax1 = plot.subplots()

# Use axes to create a pie chart
    # ax1.pie(data array, explode array, labels array, starting angle)
ax1.pie(result_bin, explode, labels, autopct='%1.1f%%', startangle=0)

# Equal aspect ratio ensures that pie is drawn as a circle.
ax1.axis('equal')

plot.show()
```

### Answer
